```python
from opacus.accountants.utils import get_noise_multiplier
n = 100
k = 50
target_epsilon = [(i+10)*0.005/(n**1.2) for i in range(k)]
target_delta = [(i+1)*0.001/(n**1.2) for i in range(k)]
noise_multipliers = [0] * k
 
for i in range(k):
    noise_multipliers[i] = get_noise_multiplier(
            target_epsilon=target_epsilon[i],
            target_delta=target_delta[i],
            sample_rate=1/(n),
            epochs=1,
            accountant='gdp',
        )
```
